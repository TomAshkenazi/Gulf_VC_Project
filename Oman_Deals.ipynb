```python
import pandas as pd
```

```python
OmanTech_File = "Oman_Deals/Oman_Tech_Fund_Deals.xlsx"
OmanTech_1 = pd.read_excel(OmanTech_File, header=7)
OmanTech = pd.DataFrame(OmanTech_1)
OmanTech = OmanTech[OmanTech['Deal ID'].notna()]
fund1="Oman Technology Fund"
OmanTech['Fund'] = fund1
OmanTech
```

```python
IDO_File = "Oman_Deals/IDO_Deals.xlsx"
IDO_1 = pd.read_excel(IDO_File, header=7)
IDO = pd.DataFrame(IDO_1)
IDO = IDO[IDO['Deal ID'].notna()]
fund2="IDO Investments"
IDO['Fund'] = fund2
IDO
```

```python
funds = [OmanTech,IDO]
all_Omani_deals = pd.concat(funds)
all_Omani_deals
```

```python
OMAN_DEALS = all_Omani_deals.reset_index(drop=True)
```

```python
OMAN_DEALS
```

```python
country ="Oman"
OMAN_DEALS["Fund's Country"] = country
OMAN_DEALS
```

```python
OMAN_DEALS['Deal Size'] = OMAN_DEALS['Post Valuation']*(OMAN_DEALS['% Acquired']/100)
OMAN_DEALS
```
